```python
import pandas as pd
import pyterrier as pt ; pt.init()
from pyterrier_pisa import PisaIndex
```

```python
index = PisaIndex.from_dataset('msmarco_passage', 'pisa_porter2')
retr = index.bm25(num_results=5)
```

```python
pipeline = retr >> pt.text.get_text(pt.get_dataset('irds:msmarco-passage'), 'text')
```

```python
pipeline(pd.DataFrame([
  {'qid': '1108939', 'query': 'what slows down the flow of blood'},
  {'qid': '1112389', 'query': 'what is the county for grand rapids, mn'},
  {'qid': '792752', 'query': 'what is ruclip'},
]))
```
